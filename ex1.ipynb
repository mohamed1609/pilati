```python
#ex 1.1 dimensionning of storage area


from docplex.mp.model import Model

mdl = Model(name ='storage_dimensioning') #de

# parameters

M = 500 #total number of request location
wx= 3.5 #dimension1 of aisle m
wy= 3.5 #dimension2 of aisle m
v = 5 #speed of forklift km/h
v = 5/3.6  #speed of forklift m/s
l = 4 #levels of storage
a = 1.05 #pallet space 
b = 1.05 #pallet space 

# decision variable

n = mdl.integer_var(name = 'n')  #number of storage location in the x dimension
m = mdl.integer_var(name = 'm')  #number of storage location in the y dimension
k = mdl.integer_var(name = 'k')  #the real decision variable 

#defining the constraints

mdl.add_constraint(n == 2*k)   #n should be even 
mdl.add_constraint(n*m*l >= M) #to satisfy the required locations number
mdl.add_constraint(m >= 0)       #m sould be positive
mdl.add_constraint(k >= 0)       #k sould be positive

#objective function

mdl.minimize((a+wx/2)*n/v+(b*m+wy)/(2*v))

print(mdl.export_to_string())

sol = mdl.solve(log_output = True )  #solving the problem

sol.display()     #display solution


n = sol.get_value(n)
m = sol.get_value(m)
k = sol.get_value(k)

print("warehouse dimensions --> lenght, width", (a+wx/2)*n, (b*m+wy))  #printing the final 2 dimensions of the warehouse
```
